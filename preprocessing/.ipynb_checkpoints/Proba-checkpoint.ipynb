```python
from scipy.io import arff
import pandas as pd
import os
import numpy as np
import missingno as msno
import matplotlib.pyplot as plt
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.ensemble import RandomForestRegressor
from colorama import Fore
import seaborn as sns
import impyute as impy
from sklearn.model_selection import train_test_split

# ucitavanje podataka 
def load_data_sets():
    ds = []
    data_path = '..'
    data_path = os.path.join(data_path, 'data')
    files = sorted(os.listdir(data_path))
    
    atr_names = [f'X{i}' for i in range(1, 65)]
    atr_names.append('class')

    for file in files:
        full_file_path = os.path.join(data_path, file)
        data = arff.loadarff(full_file_path)
        df = pd.DataFrame(data[0])
        df.columns = atr_names
        ds.append(df)
    return ds
        
        
data_sets = load_data_sets()
data_sets[0].head()
```

```python
# prebacivanje klasa u tip int
for i in range(len(data_sets)):
    column = getattr(data_sets[i], 'class')
    data_sets[i]['class'] = column.astype(np.int32)
```

```python
data = data_sets[0]
```

```python
data.fillna(data.mean(), inplace=True)
```

```python
data.info()
```

```python
from joblib import dump

X = data.drop('class', axis=1)
Y = data['class']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, stratify=Y, random_state=41)
dump((X_train, X_test, Y_train, Y_test), 'classification_data.joblib')
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier, plot_tree

params = {
    'max_depth': [2,4,6,8],
    'criterion': ['gini', 'entropy']
}
```

```python
estimator = GridSearchCV(DecisionTreeClassifier(class_weight = {0: 1, 1: 20}), param_grid=params, cv=10, scoring='f1')
estimator.fit(X_train, Y_train)
```

```python
estimator.best_score_
```

```python
estimator.best_params_
```

```python
from sklearn.metrics import classification_report, accuracy_score, f1_score, confusion_matrix
from termcolor import colored

def report(model, x, y, text = "training"):
    y_pred = model.predict(x)
    
    print(colored("Classification report for model {} on {} data".format(type(model).__name__, text), "blue"))
    print("---------------------------------------------------------------------------------")
    print(classification_report(y, y_pred))
    print("---------------------------------------------------------------------------------")
    
    print(colored("Confusion matrix for model {} on {} data ".format(type(model).__name__, text), "blue"))
    print("---------------------------------------------------------------------------------")
    print(pd.DataFrame(confusion_matrix(y, y_pred), columns=["0", "1"], index=["0", "1"]))
    print("---------------------------------------------------------------------------------")
    print("Total accuracy: ", accuracy_score(y, y_pred))
    print("F1 score: ", f1_score(y, y_pred))
```

```python
report(estimator.best_estimator_, X_train, Y_train)
```

```python
report(estimator.best_estimator_, X_test, Y_test, "test")
```
